```python
def Shuffle_Arrays(a,b):
    assert len(a) == len(b)
    p = np.random.permutation(len(a))
    return a[p], b[p]
```

```python
main_path = os.getcwd()


ImagesArrayTrain = np.empty((0,28,28))
ImagesArrayTest = np.empty((0,28,28))

LabelArrayTrain = np.empty((0))
LabelArrayTest = np.empty((0))

#=================== BOLD ===================

#Seta Direita Bold
ImagesArray = []
LabelArray = []
label = "SetaDireita"
Style = "Bold"
label_number = 3
class_path = os.path.join(main_path,"DataBase\\"+Style+"\\"+label)
base_name = label+"-"+Style+"-"

for i in range(850):
    name = base_name+str(i+1)+".jpg"
    img_path = os.path.join(class_path,name)
    imgLoaded = cv2.imread(img_path,0)
    LabelArray.append(label_number)
    ImagesArray.append(imgLoaded)

np.random.shuffle(ImagesArray)
print(np.shape(ImagesArray))
ImagesArrayTrain = np.concatenate((ImagesArrayTrain,ImagesArray[:800]))
ImagesArrayTest = np.concatenate((ImagesArrayTest,ImagesArray[800:]))
LabelArrayTrain = np.concatenate((LabelArrayTrain,LabelArray[:800]))
LabelArrayTest = np.concatenate((LabelArrayTest,LabelArray[800:]))
print(np.shape(ImagesArrayTrain))
print(np.shape(ImagesArrayTest))
print(np.shape(LabelArrayTrain))
print(np.shape(LabelArrayTest))

#Seta Baixo Bold
ImagesArray2 = []
LabelArray2 = []
label = "SetaBaixo"
Style = "Bold"
label_number = 0
class_path = os.path.join(main_path,"DataBase\\"+Style+"\\"+label)
base_name = label+"-"+Style+"-"

for i2 in range(850):
    name = base_name+str(i2+1)+".jpg"
    img_path = os.path.join(class_path,name)
    imgLoaded = cv2.imread(img_path,0)
    LabelArray2.append(label_number)
    ImagesArray2.append(imgLoaded)

np.random.shuffle(ImagesArray2)
print(np.shape(ImagesArray2))
ImagesArrayTrain = np.concatenate((ImagesArrayTrain,ImagesArray2[:800]))
ImagesArrayTest = np.concatenate((ImagesArrayTest,ImagesArray2[800:]))
LabelArrayTrain = np.concatenate((LabelArrayTrain,LabelArray2[:800]))
LabelArrayTest = np.concatenate((LabelArrayTest,LabelArray2[800:]))
print(np.shape(ImagesArrayTrain))
print(np.shape(ImagesArrayTest))
print(np.shape(LabelArrayTrain))
print(np.shape(LabelArrayTest))

#Seta Cima Bold
ImagesArray3 = []
LabelArray3 = []
label = "SetaCima"
Style = "Bold"
label_number = 1
class_path = os.path.join(main_path,"DataBase\\"+Style+"\\"+label)
base_name = label+"-"+Style+"-"

for i3 in range(850):
    name = base_name+str(i3+1)+".jpg"
    img_path = os.path.join(class_path,name)
    imgLoaded = cv2.imread(img_path,0)
    LabelArray3.append(label_number)
    ImagesArray3.append(imgLoaded)

np.random.shuffle(ImagesArray3)
print(np.shape(ImagesArray3))
ImagesArrayTrain = np.concatenate((ImagesArrayTrain,ImagesArray3[:800]))
ImagesArrayTest = np.concatenate((ImagesArrayTest,ImagesArray3[800:]))
LabelArrayTrain = np.concatenate((LabelArrayTrain,LabelArray3[:800]))
LabelArrayTest = np.concatenate((LabelArrayTest,LabelArray3[800:]))
print(np.shape(ImagesArrayTrain))
print(np.shape(ImagesArrayTest))
print(np.shape(LabelArrayTrain))
print(np.shape(LabelArrayTest))

#Seta Esquerda Bold
ImagesArray4 = []
LabelArray4 = []
label = "SetaEsquerda"
Style = "Bold"
label_number = 2
class_path = os.path.join(main_path,"DataBase\\"+Style+"\\"+label)
base_name = label+"-"+Style+"-"

for i4 in range(850):
    name = base_name+str(i4+1)+".jpg"
    img_path = os.path.join(class_path,name)
    imgLoaded = cv2.imread(img_path,0)
    LabelArray4.append(label_number)
    ImagesArray4.append(imgLoaded)

np.random.shuffle(ImagesArray4)
print(np.shape(ImagesArray4))
ImagesArrayTrain = np.concatenate((ImagesArrayTrain,ImagesArray4[:800]))
ImagesArrayTest = np.concatenate((ImagesArrayTest,ImagesArray4[800:]))
LabelArrayTrain = np.concatenate((LabelArrayTrain,LabelArray4[:800]))
LabelArrayTest = np.concatenate((LabelArrayTest,LabelArray4[800:]))
print(np.shape(ImagesArrayTrain))
print(np.shape(ImagesArrayTest))
print(np.shape(LabelArrayTrain))
print(np.shape(LabelArrayTest))

ImagensArrayTrainFinal, LabelArrayTrainFinal = Shuffle_Arrays(ImagesArrayTrain,LabelArrayTrain)
ImagensArrayTestFinal, LabelArrayTestFinal = Shuffle_Arrays(ImagesArrayTest,LabelArrayTest)

BoldTrainData = ImagensArrayTrainFinal, LabelArrayTrainFinal
BoldTestData = ImagensArrayTestFinal, LabelArrayTestFinal
```
